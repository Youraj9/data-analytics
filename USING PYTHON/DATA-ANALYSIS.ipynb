Question 2: Mid-Level Aggregation
Now that your data is clean (let's assume we saved it as a DataFrame called df_clean), your Sales Manager has come to your desk with a few questions about performance.

This will test your ability to use .groupby() and the .agg() dictionary method we talked about earlier!

Your Task:


Write the code to group the data by Region, and find the Total (sum) of the Sales for each region.

Write the code to group the data by Product, and use the .agg() function to find two things at once:

The Average (mean) of the Value column.

The Maximum (max) of the Sales column.

```python
# q1 
aggregateddata = missingvalues.groupby("Region")["Sales"].sum()
print("SUM OF THE SALES IN EACH REGION")
print(aggregateddata)
```

# THERE ARE TWO WAY OF DOING IT FOLLOWING IT IS -  > 

```python
# syntax 
# dataframe.groupby().agg({"column1":"function1","column2":"function2"})
# in the agg function we can also give the name to the new column by using 
# dataframe.groupby().agg( NEWCOLUMN1 = ("column1","function1"), NEWCOLUMN2 = ("column2","function2") )
```

```python
missingvalues.groupby("Product").agg( MEANVALUES=("Value","mean"), MAXSALES=("Sales","max"))
```

```python
#q2 
productanalysis = missingvalues[["Value","Sales","Product"]]
print(productanalysis.head(6))
# 
print(productanalysis.groupby("Product")["Sales"].mean()) # for checking the total sales are same 
print(productanalysis.groupby("Product")["Value"].max())
productanalysis.groupby("Product").agg({"Value":"mean","Sales":"max"}).rename(columns={"Value":"MEANVALUE","Sales":"MAXSALES"})
```

Question 4: Feature Engineering & Sorting (Mid-Level)
The Sales Manager is thrilled with your data pulls. Now, they want to do some actual performance analysis. Sometimes the data we need doesn't exist yet, so we have to invent it using the data we already have! This is called Feature Engineering.

Assume we are back to using our clean DataFrame (let's call it data). Let's assume the Value column actually represents the "Number of Units Sold."

Your Task:

Create a New Column: Write the code to create a brand new column in your DataFrame called Price_Per_Unit. This should be calculated by dividing the Sales column by the Value column.

Sort the Data: Write the code to sort the entire DataFrame by this new Price_Per_Unit column in descending order (meaning the highest price should be at the very top).

Show the Leaders: Write the code to display only the top 5 rows of this newly sorted DataFrame.

```python
# task four 
#creating a new column here
df=data.copy()
df["Price_Per_Unit"]=df["Sales"]/df["Value"]
print(df.head(6))

df.sort_values(by="Price_Per_Unit",ascending=False).head(5)
```
